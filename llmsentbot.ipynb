```python
def analyze_sentiment(text):
    result = sentiment_pipeline(text)[0]
    
    label = result["label"]
    score = round(result["score"], 3)
    
    if label.startswith("LABEL"):
        label_id = int(label.split("_")[1])
        label = sentiment_pipeline.model.config.id2label[label_id]

    label = label.capitalize()

    if score < 0.6:
        print("The sentiment is unclear.")
    else:
        print(f"This text expresses a {label} sentiment.")
        print(f"(Confidence: {score})")
```

```python
from transformers import pipeline

sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model="siebert/sentiment-roberta-large-english"
)

print("Model loaded successfully!")
```

```python
print(sentiment_pipeline("good"))
print(sentiment_pipeline("bad"))
print(sentiment_pipeline("amazing"))
print(sentiment_pipeline("I love this"))
print(sentiment_pipeline("I hate this"))
```

```python
def sentiment_chatbot():
    print("Sentiment Chatbot 🤖")
    print("Type 'exit' to stop.\n")

    while True:
        user_input = input("You: ")

        if user_input.lower() == "exit":
            print("Bot: Goodbye! 👋")
            break

        result = sentiment_pipeline(user_input)[0]
        label = result["label"]
        score = round(result["score"], 3)

        if label == "POSITIVE":
            print(f"Bot: This is Positive. 😊 (confidence: {score})\n")

        elif label == "NEGATIVE":
            print(f"Bot: This is Negative. 😔 (confidence: {score})\n")

sentiment_chatbot()
```

```python
import ollama

def llm_chatbot():
    print("LLM Chatbot (Open Source) 🤖")
    print("Type 'exit' to stop.\n")

    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            print("Bot: Goodbye! 👋")
            break

        resp = ollama.chat(
            model="mistral",
            messages=[{"role": "user", "content": user_input}]
        )
        print("Bot:", resp["message"]["content"])
```

```python
llm_chatbot()
```

```python
def sentiment_llm_chatbot():
    print("LLM Sentiment Chatbot 🤖")
    while True:
        text = input("You: ")
        if text.lower() == "exit":
            break

        prompt = f'''
        Answer to the query like you are a good close friend:

        "{text}"
        '''
        resp = ollama.chat(
            model="mistral",
            messages=[{"role":"user","content":prompt}]
        )

        print("Bot:", resp["message"]["content"])
```

```python
sentiment_llm_chatbot()
```
